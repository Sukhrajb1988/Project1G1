```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv 
```

```python
# Read data from Samsung Excel file
samsung_data = pd.read_csv("Samsung 2019_USD.csv")
samsung_close_prices = samsung_data['Close']
samsung_date = pd.to_datetime(samsung_data['Date'])
print(samsung_date)
```

```python
# Read data from Apple Excel file
apple_data = pd.read_csv("Apple 2019.csv")
apple_close_prices = apple_data['Close']
apple_date = pd.to_datetime(apple_data['Date'])
print(apple_date)
```

```python
#Create a line graph

plt.plot(samsung_date, samsung_close_prices, label='Samsung', color='b')
plt.plot(apple_date, apple_close_prices, label='Apple', color='r')

#Title 
plt.title('Samsung vs Apple Stock Prices 2019')

#Label
plt.xlabel('Date')
plt.ylabel('Stock Price in USD')

#legend 
plt.legend(loc = "best")

plt.grid()
plt.show()
```

```python
# Read CSV files into DataFrames
apple_stock = pd.read_csv("AAPL.csv")
samsung_stock = pd.read_csv("Samsung.csv")

# Merge based on the "Date" column
merged_stock = pd.merge(samsung_stock, apple_stock, on="Date", how="inner")

# Save the merged DataFrame to a new CSV file
merged_stock.to_csv("merged_stock.csv", index=False)

print(merged_stock)
```

```python
# Read the merged stock data
merged_stock = pd.read_csv("merged_stock.csv")

# Convert the "Date" column to datetime format
merged_stock['Date'] = pd.to_datetime(merged_stock['Date'])

# Plotting
plt.figure(figsize=(10, 6))

# Plot Apple stock price
plt.plot(merged_stock['Date'], merged_stock['Apple Close'], label='Apple', color='r')

# Plot Samsung stock price
plt.plot(merged_stock['Date'], merged_stock['Samsung Close'], label='Samsung', color='b')

# Customize the plot
plt.title('Apple and Samsung Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()

# Save the plot as an image (e.g., PNG)
plt.savefig("Apple & Saumsung stock_prices_plot_Total.png")
```

```python

# Read the merged stock data
merged_stock = pd.read_csv("merged_stock.csv")

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(merged_stock['Samsung Close'], merged_stock['Apple Close'])[0, 1]

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot to visualize the relationship
plt.scatter(merged_stock['Samsung Close'], merged_stock['Apple Close'])
plt.title('Scatter Plot: Samsung vs Apple Stock Prices')
plt.xlabel('Samsung Stock Price')
plt.ylabel('Apple Stock Price')

# Perform linear regression
slope, intercept = np.polyfit(merged_stock['Samsung Close'], merged_stock['Apple Close'], 1)

# Regression line
regression_line = slope * merged_stock['Samsung Close'] + intercept
plt.plot(merged_stock['Samsung Close'], regression_line, color='red', label='Regression Line')


# Show the correlation coefficient in the plot
plt.text(plt.xlim()[0], plt.ylim()[1], f'Correlation Coefficient: {correlation_coefficient:.2f}', ha='left', va='top', fontsize=10, color='red')

# Show the plot
plt.show()
```
